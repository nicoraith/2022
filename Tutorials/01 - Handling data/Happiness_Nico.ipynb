## Data from World Happiness Report

The World Happiness Report is an annual publication of the United Nations Sustainable Development Solutions Network. It contains articles, and rankings of national happiness based on respondent ratings of their own lives, which the report also correlates with various life factors.

In this notebook we will explore the happiness of different countries and the features associated.
The datasets that we will use are available in *Data*: **happiness2020.pkl** and **countries_info.csv**.

Although the features are self-explanatory, here a summary: 

**happiness2020.pkl**
* country: *Name of the country*
* happiness_score: *Happiness score*
* social_support: *Social support (mitigation the effects of inequality)*
* healthy_life_expectancy: *Healthy Life Expectancy*
* freedom_of_choices: *Freedom to make life choices*
* generosity: *Generosity (charity, volunteers)*
* perception_of_corruption: *Corruption Perception*
* world_region: *Area of the world of the country*

**countries_info.csv**
* country_name: *Name of the country*
* area: *Area in sq mi*
* population: *Number of people*
* literacy: *Literacy percentage*

```python
import pandas as pd
%matplotlib inline

DATA_FOLDER = 'Data/'

HAPPINESS_DATASET = DATA_FOLDER+"happiness2020.csv"
COUNTRIES_DATASET = DATA_FOLDER+"countries_info.csv"
```

## Task 1: Load the data

Load the 2 datasets in Pandas dataframes (called *happiness* and *countries*), and show the first rows.


**Hint**: Use the correct reader and verify the data has the expected format.

```python
happiness = pd.read_csv(HAPPINESS_DATASET, sep=',')
happiness.head()
```

```python
countries = pd.read_csv(COUNTRIES_DATASET, sep=',')
countries.head()
```

## Task 2: Let's merge the data

Create a dataframe called *country_features* by merging *happiness* and *countries*. A row of this dataframe must describe all the features that we have about a country.

**Hint**: Verify to have all the rows in the final dataframe

```python
# add a new column in countries called 'country' that takes the countries in 'country_name' and puts their first letter in uppercase
#countries['country'] = 
countries['country'] = countries['country_name'].apply(lambda a:a.title())
countries.head(50)
```

```python
happiness.head()
```

```python
# merge the two data frames by the 'country' column
# as a result, two 'country_name' columns appear
# thus we have to delete the two, namely: 'country_name_x' and 'country_name_y'
# we put the axis=1 to drop column-wise
country_features = happiness.merge(countries, on='country')
country_features = country_features.drop(['country_name'], axis=1)
country_features.head()
```

```python
print("happiness = {}".format(len(happiness)))
print("countries = {}".format(len(countries)))
print("country_features = {}".format(len(country_features)))

len(happiness) == len(countries) == len(country_features)
```

## Task 3: Where do people are happier?

Print the top 10 countries based on their happiness score (high is better).

```python
country_features[['country','happiness_score']].sort_values(ascending=False,by='happiness_score').head(10)
```

We are interested to know in what world region the people are happier. 

Create and print a dataframe with the (1) average happiness score and (2) the number of contries for each world region.
Sort the result to show the happiness ranking.

```python
# we want to create a new df with countries grouped by their regions, we want to count how many there are in each region
# we want the average happiness score in every region
# group by 'wrold_region' over happiness_score and we take the mean and count them
countries_by_region = country_features.groupby("world_region")['happiness_score'].agg(['mean','count']) # we need to specify what label we want to keep
countries_by_region.sort_values('mean',ascending=False)
```

The first region has only a few countries! What are them and what is their score?

```python
country_features[country_features.world_region == 'North America and ANZ'][['country', 'happiness_score']].head(20)
```

## Task 4: How literate is the world?

Print the name of countries with a level of literacy of 100%. 

For each country, print the name and the world region with the format: *{region name} - {country name} ({happiness score})*

```python
country_features[country_features.literacy == 100][['world_region','country', 'happiness_score']].head(20)
```

What is the global average?

```python
# replacing commas by dots, then we need to comment the line because the change has already happened
#country_features['literacy'] = country_features['literacy'].str.replace(',','.')

# we convert the string to int
country_features['literacy'] = pd.to_numeric(country_features['literacy'])

# computing the mean
country_features.literacy.mean()
country_features.head()
```

Calculate the proportion of countries with a literacy level below 50%. Print the value in percentage, formatted with 2 decimals.

```python
low_literacy = country_features[country_features.literacy <= 50].country.count()
total_countries = country_features.country.count()
prop_low_literacy = 100*low_literacy/total_countries

print("Percentage of countries with literacy level below 50%:",  prop_low_literacy)
```

Print the raw number and the percentage of world population that is illiterate.

```python
total_iliterate_bycountry = country_features['population']*((100-country_features['literacy'])/100)
total_iliterate_bycountry.sum()
country_features.population.sum()
print("The total number of illiterate people in the world is: {:0.0f}".format(total_iliterate_bycountry.sum()))
print("The percentage of the world population that is illiterate is: {:0.2f}".format(100*total_iliterate_bycountry.sum()/country_features.population.sum()), "%")
```

## Task 5: Population density

Add to the dataframe a new field called *population_density* computed by dividing *population* by *area*.

```python
country_features.dtypes
country_features['population_density'] = country_features['population']/country_features['area']
country_features.head()
```

What is the happiness score of the 3 countries with lowest population density?

```python
# always when you want to display different columns you write how you want to sort it and then [['column 1', 'column 2', ...]]

country_features.sort_values('population_density', ascending=True)[['country', 'happiness_score', 'population_density']].head(3)
```

Plot in scatter plot the happiness score (x) and healty life expectancy (y).
