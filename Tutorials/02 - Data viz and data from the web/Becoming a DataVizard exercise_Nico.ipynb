```python
# This is a helper bootstrapping function you can use in this exercise to obtain 95% confidence
# intervals around the estimated average. The underlying concept will be explained in the next week's lecture.

# Input: your array and the number of random samples (e.g., 1000 is a good number)
# Output: [lower error, upper error]

def bootstrap_CI(data, nbr_draws):
    means = np.zeros(nbr_draws)
    data = np.array(data)

    for n in range(nbr_draws):
        indices = np.random.randint(0, len(data), len(data))
        data_tmp = data[indices] 
        means[n] = np.nanmean(data_tmp)

    return [np.nanpercentile(means, 2.5),np.nanpercentile(means, 97.5)]
```

```python
# Calculate yearly stats across movies here

# We want to group the movies by year, then add up all the years and get the average of revenue per year
# but we also want to compute the confidence interval of each year


# def bootstrap_CI(data, nbr_draws):
#     means = np.zeros(nbr_draws)
#     data = np.array(data)

#     for n in range(nbr_draws):
#         indices = np.random.randint(0, len(data), len(data))
#         data_tmp = data[indices] 
#         means[n] = np.nanmean(data_tmp)

#     return [np.nanpercentile(means, 2.5),np.nanpercentile(means, 97.5)]


# grouped_year = movies.groupby(['year'])   # here we group by year then we create a series with three columns and we give values to the cells
statistics = movies.groupby(['year']).apply(lambda x: pd.Series({    
    'worldwide_yearly_average':x['worldwide_gross'].mean(),
    'lower_bound_error': bootstrap_CI(x['worldwide_gross'], 1000)[0],
    'upper_bound_error': bootstrap_CI(x['worldwide_gross'], 1000)[1]
})) 
statistics.head()
```

```python
# Write your code to make plot with errorbars here

# Confidence interval: CI = mean +- lower or upper bound
plt.title('Yearly Average Revenue')
plt.xlabel('Year')
plt.ylabel('Average Revenue [$]')
plt.plot(statistics.index, statistics['worldwide_yearly_average'])
plt.errorbar(statistics.index, statistics['worldwide_yearly_average'], 
             yerr = [-statistics['lower_bound_error'] + statistics['worldwide_yearly_average'], 
                     -statistics['worldwide_yearly_average'] + statistics['upper_bound_error']], fmt ='c')

# lower error: mean - lower values
# upper error: upper values - mean
plt.show()
```

```python
# Write your code to make filled plot here

plt.fill_between(statistics.index, statistics.lower_bound_error, statistics.upper_bound_error, color = 'gray')
plt.plot(statistics.index, statistics.worldwide_yearly_average, color = 'red')
```

```python
# Calculate stats by genre here

# here we want the duration of each genre so we do like previously
# we create a series with the values of the length of the movies

movies_by_duration = movies.groupby(['Main_Genre']).apply(lambda x: pd.Series({'length': x['length'].values}))
# we applied a function with for us is to create a new table with the genres as INDEX and the values of the lengths of the movies 
# for the attribute length
movies_by_duration.head(16)
```

```python
# Write your code to make 4x4 panel here

fig, ax = plt.subplots(4,4, figsize = (8,6), sharey = True, sharex = True) ## we initialise a figure with 4 by 4 axes

for i in range (16):
    subplt = ax[i%4, math.floor(i/4)] ## we give the positions of each subplot 0,0; 0,1; 0,2; 0,3
    subplt.hist(movies_by_duration.iloc[i].values) ## we use iloc because we want to access integer numbers while the loc function allows to access strings
    subplt.set_title(movies_by_duration.index[i]) ## function set title and we get as title the index of the dataframe which corresponds to a movie genre
    
fig.tight_layout()
```

```python
# Write your code to make the first heatmap here

# Use seaborn heatmap function
# We want to obtain the frequencies of each genre by studio
# For this we will use the panda function crosstab: Compute a simple cross tabulation of two (or more) factors.
# By default, computes a frequency table of the factors unless an array of values and an aggregation function are passed.

df_genre_studio = pd.crosstab(movies['Main_Genre'], movies['studio'])
sns.heatmap(df_genre_studio)
```
